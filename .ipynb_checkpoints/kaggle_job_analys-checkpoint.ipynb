### Salary Estimate 
1. It hsa a missing value -> find out why and what can I do with it
2. Str format of salary Estimate -> I want to write at least 3 new columns (min max avg) + less or greater than average salary in the US -> then count + highlight high end jobs 

```python
# Salary Estiamte has 1 missing value
df['Salary Estimate'].isnull().sum()
```

```python
# row #2149 has a missing Salary Estimate - the question is: are there any other advertisments from this company
df[df['Salary Estimate'].isnull()]
```

```python
# these jobs are not the same, and thus we are not going to include 2149
df[df['Company Name'] == 'Protingent\n4.4']
```

```python
df.drop(2149, inplace = True)
```

```python
df['Salary Estimate'].isnull().sum()
```

```python
#Another problem with job salary estimates is that it is written in str format -> we need to change it to int format
# There is a min - max - average
df['Salary Estimate'].sample(10)
```

```python
def min_salary(value):
    return value['Salary Estimate'].str.lstrip('$').str[:3].str.replace('K', '').str.strip().astype('float')
def max_salary(value):
    return value['Salary Estimate'].str[6:10].str.replace('K','').str.lstrip('$').str.strip().astype('float')

df.insert(3, 'Salary_min', min_salary(df))
df.insert(4, 'Salary_max', max_salary(df))
df.insert(5, 'Salary_average', df[['Salary_min', 'Salary_max']].mean(axis=1))
```

```python
# the average anuual income in the US is 52
across_country = np.where(np.array(df['Salary_average']) > 52, 'greater than average' , 'less than average')
df.insert(6, 'Across_country', across_country)
```

```python
# pretty interesting 
df['Across_country'].value_counts()
```

```python
df['Salary_min'].isnull().any()
```

```python
df['Salary_max'].isnull().any()
```

```python
new_df_min = pd.DataFrame(df_a['Salary_min'])
new_df_min['Labels'] = 'Salary_min'
new_df_min = new_df_min.rename(columns = {'Salary_min' : 'Salary'})

new_df_max = pd.DataFrame(df_a['Salary_max'])
new_df_max['Labels'] = 'Salary_max'
new_df_max = new_df_max.rename(columns = {'Salary_max' : 'Salary'})

new_df = new_df_min.append(new_df_max, ignore_index = True)

px.histogram(new_df, x="Salary", color="Labels", barmode="overlay", title = 'Min Salary and Max Salary distribution')
```

```python
new_df = pd.DataFrame(df_a['Job Title'].value_counts().head(15))
new_df = new_df.rename(columns = {'Job Title' : 'Most Popular'})
new_df['Job Title'] = new_df.index
new_df = new_df.set_index(np.arange(15))
```

```python
px.line(new_df, x='Job Title', y='Most Popular', markers = True, title = '15 most in demand jobs in data field')
```
